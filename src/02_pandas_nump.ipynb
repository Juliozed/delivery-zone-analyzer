## Flag, new columns for alerts

```python
df["delivery_status"] = np.where(
        df["days_late"] > 0, "Late", 
        np.where(df["days_late"] == 0, "On Time", "Early"))
```

```python
df["delivery_status"].value_counts()
```

```python
df.info()
```

```python
df["freight_ratio"] = round((df["freight"]/ df["order_value"]) * 100, 2)
```

```python
df["freight_ratio"] = round((df["freight"]/ df["order_value"].replace(0, np.nan)) * 100,2)
```

```python
df[["order_id", "status", "freight_ratio"]].head(10)
```

```python
df["freight_ratio"] = df["freight_ratio"].fillna(0)
df[["freight_ratio", "status", "order_id"]].head(10)
```

```python
df[["order_id", "city", "status", "days_late", "delivery_status", "freight_ratio"]].head(10)
```

```python
df.groupby("city")["delivery_cost"].mean().round(2).sort_values(ascending=False)
```

1. Average days_late by courier — which courier is worst?

```python
df.groupby("courier")["days_late"].mean().round().sort_values(ascending=False)
```

2. Total order_value by state — which state generates most revenue?

```python
df.groupby("state")["order_value"].sum().sort_values(ascending=False)
```

3. Count of orders by delivery_status — you already did this with value_counts() but try it with groupby this time using .count()

```python
df.columns
```

```python
df.groupby("delivery_status")["order_id"].count()
```

```python
df.groupby(["state", "courier"])["days_late"].mean().round(2).sort_values(ascending=False)
```

```python
df.groupby("courier").agg(
    total_orders = ("order_id", "count"),
    avg_days_late = ("days_late", "mean"),
    late_count = ("days_late", lambda x: (x > 0).sum()),
).round(2)
```

```python
df["cost_category"] = np.where(df["delivery_cost"] > 20, "High",
                    np.where(df["delivery_cost"] > 10, "Medium", "Low")
    )
```

```python
df[["order_id", "delivery_cost", "cost_category"]].head(20)
```

```python
df.columns
```

```python
df["breach_severity"] = np.where(df["days_late"] > 7, "Critical",
                  np.where(df["days_late"] > 3, "Moderate",
                  np.where(df["days_late"] > 0, "Minor", "None")))
```

```python
df[["order_id", "status", "breach_severity"]].head(20)
```

```python
df["breach_severity"].value_counts()
```

```python
df["region"] = np.where(df["state"].isin(["CA", "OR", "WA"]), "West",
                np.where(df["state"].isin(["NY", "MA", "PA", "MD"]), "Northeast", 
                np.where(df["state"].isin(["TX", "FL", "GA", "NC", "TN", "LA"]), "South", "Midwest" 
)))
```

```python
df["region"].value_counts()
```

```python
df["order_value_with_tax"] = df["order_value"].apply(lambda x: round(x + (x * 0.085),2))
df[["order_id", "order_value", "order_value_with_tax"]].head(20)
```

```python
df["delivery_cost_per_kilogram"] = df.apply(lambda row: round(row["delivery_cost"] / row["weight_kg"],2), axis=1)
```

```python
df[["order_id", "delivery_cost", "weight_kg",  "delivery_cost_per_kilogram", "status"]].head(20)
```

```python
df.groupby("region").agg(
    total_orders = ("order_id", "count"),
    avg_delivery_cost = ("delivery_cost", "mean"),
    avg_days_late = ("days_late", "mean"),
    late_orders = ("days_late", lambda x: (x > 0).sum()),
    total_revenue = ("order_value", "sum"),
    courier = ("courier", "max")
).round(2).sort_values("avg_days_late", ascending=False)
```

```python
df[df["region"].isin(["Midwest","South",])].groupby("courier").agg(
    total_orders = ("order_id", "count"),
    avg_days_late = ("days_late", "mean"),
    late_orders = ("days_late", lambda x: (x > 0).sum())
).round(2).sort_values("avg_days_late", ascending=False)
```

```python
df[df["region"].isin(["Northeast","West",])].groupby("courier").agg(
    total_orders = ("order_id", "count"),
    avg_days_late = ("days_late", "mean"),
    late_orders = ("days_late", lambda x: (x > 0).sum())
).round(2).sort_values("avg_days_late", ascending=False)
```

## export to excel


```python
summary = df.groupby(["region", "courier"]).agg(
    total_orders = ("order_id", "count"),
    avg_days_late = ("days_late", "mean"),
    late_orders = ("days_late", lambda x: (x > 0).sum()),
    total_revenue = ("order_value", "sum")
).round(2).reset_index()

```

```python
summary.to_excel("courier_performance_report.xlsx", index=False)
```
