```python
def get_avg_cost(orders):
    total = 0
    count = 0
    
    for order in orders:
        if order["status"] == 'delivered':
            total += order["delivery_cost"]
            count += 1  
            
        if count == 0:
            return 0
            return total / count
    return total/count
print(get_avg_cost(orders))




```

```python
def get_late_rate(orders):
    late_count = 0
    delivered_count = 0

    for order in orders:
        if order["status"] == 'delivered':
            delivered_count += 1
            if order["days_late"] > 0:
                late_count += 1
    if delivered_count == 0:
        return 0        
    return (late_count / delivered_count) * 100

print(get_late_rate(orders))
```

🚨 KPI ALERT: Freight Ratio <br>
Formula: (Total Freight ÷ Total Order Value) × 100 <br>
Result: 5.4% — freight consumes 5.4% of revenue on average <br>
What it means: For every 100 pesos in orders, 5.40 pesos goes to shipping <br>
Red flag threshold: Anything above 15% signals the company is losing margin on delivery <br>
Business decision: Zones with high freight ratio need minimum order values, route optimization, or renegotiated <br> 
courier contracts <br>
Who tracks it: Finance analysts, logistics analysts, operations managers <br>

```python
def get_freight_ratio(orders):
    total_freight = 0
    total_order_values = 0 
    
    for order in orders:
        if order["status"] == 'delivered':
            total_order_values += order["order_value"]
            total_freight += order["freight"]
    if total_order_values == 0:
        return 0
    return (total_freight/total_order_values) * 100

print(get_freight_ratio(orders))
    
```

```python
def zone_summary(city_name, orders):
    city_orders = []
    
    for order in orders:
        if order["city"] == city_name:
            city_orders.append(order)
            
    print("Zone:",city_name)
    print("Avg Cost:", round(get_avg_cost(city_orders),2))
    print("Late Rate:", round(get_late_rate(city_orders),2))
    print("Freight Ratio: ", round(get_freight_ratio(city_orders),2))
zone_summary("Manaus", orders)
zone_summary("São Paulo", orders)
zone_summary("Fortaleza", orders)
```

```python
cities = set()

for order in orders:
    cities.add(order["city"])
    
for city in cities:
    zone_summary(city, orders)
        
```
