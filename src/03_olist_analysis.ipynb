## KPIs
- delivery status
- breach severity

- Each one becomes a column in df_master that you can then use in groupby, filters, and Power BI slicers

- Since we don't have a days_late column, we should create one, start thinking like an analyst!
- we don't have it, well make it!
  

```python
df_master["days_late"] = (df_master["order_delivered_customer_date"] - df_master["order_estimated_delivery_date"]).dt.days
df_master["days_late"].describe()
```

```python
df_master["delivery_status"] = np.where(
df_master["days_late"] > 0, "Late",
np.where(df_master["days_late"] == 0, "On Time", "Early"
))
```

```python
df_master["delivery_status"].value_counts()
```

```python
df_master["breach_severity"] = np.where(df_master["days_late"] > 7, "Critical",
                               np.where(df_master["days_late"] > 3, "Moderate",
                               np.where(df_master["days_late"] > 0, "Minor", "None")))
df_master["breach_severity"].value_counts()
```

```python
df_master["freight_ratio"] = np.where(
    df_master["payment_value"] > 0,
    round((df_master["freight_value"] / df_master["payment_value"]) * 100, 2),
    0
)
```

```python
df_master["freight_ratio"].describe()
```

## whenever we get a max of 293,000% we flag and cap
- cap the freight ratio to 100%, some are mere cents and it throws the data off
- see the solution below

```python
# flag suspicious rows
df_master["flag_freight"] = np.where(df_master["freight_ratio"] > 100, "Review", "Ok")

# cap the ratio at 100%
df_master["freight_ratio"] = np.where(df_master["freight_ratio"] > 100, 100, df_master["freight_ratio"])
```

```python
df_master["flag_freight"].value_counts()
```
