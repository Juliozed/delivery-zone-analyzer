## reviews

```python
df_order_reviews["review_creation_date"]     = pd.to_datetime(df_order_reviews["review_creation_date"])
df_order_reviews["review_answer_timestamp"]  = pd.to_datetime(df_order_reviews["review_answer_timestamp"]) 
 
df_order_reviews_clean = df_order_reviews.copy()
quick_check(df_order_reviews_clean)
```

## products


```python
df_products_clean = df_products.dropna(subset=["product_category_name"]).copy()
df_products_clean = df_products_clean.rename({
    "product_name_lenght": "product_name_length",
    "product_description_lenght": "product_description_length"
})
quick_check(df_products_clean)
```

```python
df_sellers["seller_zip_code_prefix"] = df_sellers["seller_zip_code_prefix"].astype(str)

df_sellers_clean = df_sellers.copy()
quick_check(df_sellers_clean)
```

```python
df_master = df_orders_clean.merge(df_customers_clean, on="customer_id", how="left")
quick_check(df_master)
```

```python
df_master = df_master.merge(df_order_items_clean, on="order_id", how="left")
quick_check(df_master)
```

```python
df_master = df_master.merge(df_order_payments_clean, on="order_id", how="left")
print(df_master.shape)
```

```python
df_master = df_master.merge(df_products_clean, on="product_id", how="left")
print(df_master.shape)
```

```python
df_master = df_master.merge(df_sellers_clean, on="seller_id", how="left")
print(df_master.shape)
```

```python
quick_check(df_master)
```

```python
text_cols = ["product_category_name"]
numeric_cols = ["product_name_lenght", "product_description_lenght", 
                "product_photos_qty", "product_weight_g", 
                "product_length_cm", "product_height_cm", "product_width_cm"]

df_master["product_category_name"] = df_master["product_category_name"].fillna("Unknown")
for col in numeric_cols:
    df_master[col] = df_master[col].fillna(0)
```
